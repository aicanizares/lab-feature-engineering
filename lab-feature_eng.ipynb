```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

```python
spaceship.shape
```

```python
spaceship.dtypes
```

**Check for missing values**

```python
spaceship.isna().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship = spaceship.dropna()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
def extract_first_letter(value):
    return value[0]
```

```python
spaceship["Cabin"] = spaceship["Cabin"].apply(extract_first_letter)
```

```python
spaceship
```

- Drop PassengerId and Name

```python
spaceship = spaceship.drop(["PassengerId", "Name"], axis=1)
```

```python
spaceship
```

- For non-numerical columns, do dummies.

```python
spaceship_dummies = pd.get_dummies(spaceship[["HomePlanet", "CryoSleep", "Cabin", "Destination"]])
```

```python
spaceship_dummies
```

**Perform Train Test Split**

```python
features_numerical = spaceship.select_dtypes(include=["number"])
```

```python
features_numerical
```

```python
# Concatenate FEATURES (merge dfs "features_numerical" and "spaceship_dummies")
```

```python
features = pd.concat([features_numerical, spaceship_dummies], axis=1)
```

```python
features
```

```python
target = spaceship["Transported"]
```

```python
target
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
normalizer = MinMaxScaler()
```

```python
normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train
```

```python
X_train_norm
```

```python
X_train.describe()
```

```python
X_train_norm.describe()
```

```python
# Convert NumPy array to DataFrame
X_train_norm_df = pd.DataFrame(X_train_norm)
```

```python
X_train_norm.describe()
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier() 
```

```python
knn
```

```python
knn.fit(X_train_norm, y_train)
```

```python
knn.score(X_test_norm, y_test)
```

- Evaluate your model's performance. Comment it
